```python
import findspark 
findspark.init()
from pyspark import SparkContext, SparkConf
conf = SparkConf().setAppName("actions").setMaster("local[*]")
sc = SparkContext(conf = conf)
```

```python
#créer un RDD à partir du fichier text_ex1.txt
textRdd=sc.textFile('text_ex1.txt')
```

```python
# récupérer les words de chaque ligne
wordRdd=textRdd.flatMap(lambda x: x.split(" "))
```

```python
# associer chaque word avec un poids 1
wordCoupleRdd = wordRdd.map(lambda word: (word, 1))
```

```python
# regrouper les couples par words 
wordCounts = wordCoupleRdd.reduceByKey(lambda a, b: (a + b))
```

```python
# Calculer la fréquence de chaque word: 
# frequence=word[i].count/nbWords
nbWord=wordRdd.count()
wordFrequency=wordCounts.mapValues(lambda a: a/nbWord)
```

```python
#afficher le résulat
wordFrequency.collect()
```
